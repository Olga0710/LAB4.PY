<br><b>Варіант 7</b>  -   стрічка Мебіуса

```python
import numpy as np
import matplotlib.pyplot as plt
```

Визначаємо параметри поверхні та створюємо сітку для векторних обчислень

```python
u = np.linspace(0, 2 * np.pi, 60) 
v = np.linspace(-1, 1, 20)   
U, V = np.meshgrid(u, v)
```

Параметричне рівняння стрічки Мебіуса

```python
X = (1 + (V / 2) * np.cos(U / 2)) * np.cos(U)
Y = (1 + (V / 2) * np.cos(U / 2)) * np.sin(U)
Z = (V / 2) * np.sin(U / 2)
```

Відображаємо результат

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

wire = ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1, 
                         color='royalblue', linewidth=0.7, alpha=0.8)

eq_text = r'$x(u,v) = \left(1+\frac{v}{2}\cos\frac{u}{2}\right)\cos u$' + '\n' + \
          r'$y(u,v) = \left(1+\frac{v}{2}\cos\frac{u}{2}\right)\sin u$' + '\n' + \
          r'$z(u,v) = \frac{v}{2}\sin\frac{u}{2}$'

ax.plot([], [], [], label=eq_text, linestyle='none')
ax.legend(loc='upper left', fontsize=8, framealpha=0.7, title="Рівняння поверхні:")

ax.set_title('Стрічка Мебіуса', fontsize=14, fontweight='bold')

ax.set_xlabel('X', fontsize=10)
ax.set_ylabel('Y', fontsize=10)
ax.set_zlabel('Z', fontsize=10)

ax.view_init(elev=50, azim=-60)
plt.tight_layout()
plt.savefig('plot_task5_mobius_strip.png', dpi=150, bbox_inches='tight')
plt.show()
```
