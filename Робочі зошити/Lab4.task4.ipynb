<b>Задача 4.</b> Створити статичне зображення, на якому намалювати криву та супутні елементи (див. посилання). Параметри (якщо наявні) задає користувач. Якщо за посиланням зображення динамічне, то захопити його в якийсь момент часу (ненульовий). Використання циклів заборонено.
<br><b>Варіант 3</b>  -  трисектриса Маклорена

```python
import numpy as np
import matplotlib.pyplot as plt
```

Отримуємо параметр 'a' від користувача.

```python
a = float(input("Введіть значення параметра a (a > 0): "))
```

Генеруємо масив кутів theta від -pi/2 до pi/2.

```python
theta = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 1000)
```

Розрахунок полярного радіуса за формулою

```python
r = (a / 2) * (4 * np.cos(theta) - 1 / np.cos(theta))
```

Перехід від полярних координат (r, theta) до декартових (x, y).

```python
x = r * np.cos(theta)
y = r * np.sin(theta)
```

Відображаємо результат

```python
plt.figure(figsize=(10, 7))
plt.plot(x, y, label=f'Трисектриса Маклорена ($a={a}$)', color='black', linewidth=2)

plt.axvline(-a/2, color='red', linestyle='--', label='Асимптота ($x = -a/2$)')
plt.plot(0, 0, 'ro', markersize=8, label='Полюс (0,0)')
plt.plot(a, 0, 'ro', markersize=8, label=f'Полюс ({a},0)')

plt.plot(3*a/2, 0, 'gs', markersize=8, label=f'Перетин: ({3*a/2:.2f}, 0)')

plt.title('Трисектриса Маклорена', fontsize=14)
plt.axhline(0, color='darkblue', linewidth=0.8)
plt.axvline(0, color='darkblue', linewidth=0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.axis('equal') 
plt.gca().set_adjustable("box")
plt.xlim(-a*0.8, a*2)
plt.ylim(-a*1.2, a*1.2)

plt.savefig('trysektrysa_maklorena_polar.png', dpi=150)
plt.show()
```
