<b>Задача 3.</b> У завданні 4 лабораторної роботи №2 оброблялися значення функції у точках. Побудувати гістограму, яка відображає розподіл точок по інтервалах.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('lab2_var102.csv')
```

```python
t_values = raw['t']
```

```python
plt.figure(figsize=(10, 6))

plt.hist(t_values, bins=15, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Розподіл точок за інтервалами часу', fontsize=14, fontweight='bold')
plt.xlabel('Час роботи (t)', fontsize=12)
plt.ylabel('Кількість точок в інтервалі', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('plot_task3_histogram.png', bbox_inches='tight')
plt.show()
```
